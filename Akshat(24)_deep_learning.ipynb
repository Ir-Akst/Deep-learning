create tensors


```python
a=tf.constant(5)
print(a)

v=tf.constant([1,2,3,])
print(v)

m=tf.constant([[1,2],[3,4]])
print(m)

t3=tf.constant([[[1,2],[3,4]],[[4,5],[6,7]]])
print(t3)
```

tensor addition

```python
a=tf.constant([10,20,30])
b=tf.constant([50,60,70])
c=tf.add(a,b)
print(c)

d=tf.subtract(a,b)
print(d)

e=tf.multiply(a,b)
print(e)

f=tf.divide(a,b)
print(f)

g=tf.pow(a,2)
print(g)
```

```python
a=tf.constant([10,20,30])
print(a.dtype, a.shape, tf.rank(a))
```

numpy array to tensors

```python
import numpy as np
np_array = ([[1,2,3],[4,5,6]])

tensor=tf.convert_to_tensor(np_array)
print(tensor)
```
